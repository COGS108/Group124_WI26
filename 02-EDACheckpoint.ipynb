## Background and Prior Work

High-speed internet access has become a key input to modern labor markets: it enables online job search and matching, remote work, access to training, and participation in digitally mediated services (e.g., gig work, online freelancing). Yet access and adoption are uneven across the United States, creating a persistent “digital divide” that often maps onto geography (rural vs. urban), income, and education. For this project, we ask whether household broadband subscription is associated with employment and earnings for U.S. adults ages 25–54, and whether the relationship differs in rural versus urban contexts after controlling for demographic and human-capital factors.

A large body of prior research suggests that broadband can affect labor outcomes, but the direction and magnitude may depend on worker characteristics and local context. Using U.S. data from 1999–2007, Atasoy (2013) finds that broadband expansion is associated with improved labor-market outcomes, consistent with the idea that connectivity reduces frictions in job search and supports labor-force participation.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

More causal work focusing on specific subpopulations also finds meaningful effects: Dettling (2017) uses an instrumental-variables strategy based on supply-side constraints to broadband access and reports that exogenous increases in high-speed internet use raise labor force participation among married women, highlighting a plausible mechanism where internet access expands feasible work arrangements and job search at home.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Beyond participation, broadband can also influence who benefits in the labor market. Akerman, Gaarder, and Mogstad (2015) provide evidence of “skill complementarity,” where broadband adoption improves outcomes for more-skilled workers and can disadvantage less-skilled workers—suggesting that broadband may amplify existing inequalities unless paired with complementary skills and opportunities.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Recent research also emphasizes that the broadband–employment relationship may look different in rural areas, where infrastructure, adoption, and job opportunities differ from urban labor markets. For example, Isley (2022) studies rural U.S. counties during the early COVID-19 period and finds that both broadband availability and adoption are related to employment rates, using a two-stage least squares approach to address endogeneity.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

This motivates our rural–urban heterogeneity component: if broadband supports telework, job search, and access to wider labor markets, the payoff may be larger where geographic isolation is a bigger constraint; alternatively, if high-quality jobs and complementary skills are concentrated in urban areas, the earnings gains from broadband adoption may be larger in cities. Together, these studies justify examining (1) the overall association between household broadband subscription and employment/earnings and (2) whether that association differs across rural versus urban settings after accounting for education, age, race/ethnicity, household composition, and place-based factors.

Finally, prior work highlights an important methodological challenge: broadband adoption is not randomly assigned. Households with broadband may differ systematically in income, education, occupation, and local labor conditions, all of which also affect employment and earnings. The literature therefore often emphasizes careful adjustment strategies (e.g., rich controls, fixed effects, quasi-experimental instruments, or matching) to reduce confounding and clarify interpretation.<a name="cite_ref-2b"></a>[<sup>2</sup>](#cite_note-2) In this project, using public microdata, we will treat results primarily as associational unless we can justify a credible identification strategy.

## Shortcomings

This dataset has several limitations:

1. **Self-report bias**  
   Internet usage behavior and income information are self-reported by respondents, which may introduce measurement errors.

2. **Limited population coverage for some variables**  
   Some variables apply only to specific groups (for example, hourly earnings only apply to employed individuals), which may lead to missing or special coded values.

3. **Cross-sectional data**  
   The dataset reflects information from a single point in time, which limits the ability to infer causal relationships.

4. **Internet usage variables are categorical**  
   Internet usage variables mainly indicate whether the internet is used in certain locations rather than precise usage time or frequency.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Load the exact cleaned data used for this revised checkpoint
DATA_PATH = r"data/00-raw/nov23pub_cleaned.csv"
df = pd.read_csv(DATA_PATH)

print("Loaded:", DATA_PATH)
print("Shape:", df.shape)
print("Columns:", list(df.columns))
```

```python
# Prepare analysis copy
df_eda = df.copy()

# Keep the variables needed for the main EDA section
key_cols = [
    "age", "internet_at_home", "metro_status", "labor_force_status_major",
    "hourly_earnings", "PWSSWGT", "sex", "education", "race", "marital_status"
]
print("Key analysis columns present:")
print([c for c in key_cols if c in df_eda.columns])

# Metro labels used throughout the EDA
metro_map = {1: "Metro / urban", 2: "Non-metro / rural"}
df_eda = df_eda[df_eda["metro_status"].isin([1, 2])].copy()
df_eda["metro_group"] = df_eda["metro_status"].map(metro_map)

# Broadband proxy
df_eda = df_eda[df_eda["internet_at_home"].isin([0, 1])].copy()

# Employment indicator
df_eda = df_eda[df_eda["labor_force_status_major"] != -1].copy()
df_eda["employed"] = df_eda["labor_force_status_major"].isin([1, 2]).astype(int)

# Clean hourly earnings
df_eda["hourly_earnings_clean"] = pd.to_numeric(df_eda["hourly_earnings"], errors="coerce")
df_eda.loc[df_eda["hourly_earnings_clean"] <= 0, "hourly_earnings_clean"] = np.nan

df_eda["log_hourly_earnings"] = np.log(
    df_eda["hourly_earnings_clean"].clip(lower=1)
)

print("\nAnalysis sample shape:", df_eda.shape)
print("Age range:", int(df_eda["age"].min()), "to", int(df_eda["age"].max()))
print("Metro groups:", df_eda["metro_group"].value_counts().to_dict())
```

```python
# Weighted helper functions
def weighted_mean(x, w):
    x = np.asarray(x, dtype=float)
    w = np.asarray(w, dtype=float)
    m = np.isfinite(x) & np.isfinite(w)
    if m.sum() == 0 or w[m].sum() == 0:
        return np.nan
    return (x[m] * w[m]).sum() / w[m].sum()

def weighted_share(x, w):
    return weighted_mean(x, w)

def weighted_quantile(values, quantile, sample_weight):
    values = np.asarray(values, dtype=float)
    sample_weight = np.asarray(sample_weight, dtype=float)
    m = np.isfinite(values) & np.isfinite(sample_weight)
    values = values[m]
    sample_weight = sample_weight[m]
    if len(values) == 0 or sample_weight.sum() == 0:
        return np.nan
    sorter = np.argsort(values)
    values = values[sorter]
    sample_weight = sample_weight[sorter]
    cum_w = np.cumsum(sample_weight)
    cutoff = quantile * sample_weight.sum()
    return values[np.searchsorted(cum_w, cutoff, side="left")]

def grouped_weighted_share(df_in, group_cols, value_col, weight_col="PWSSWGT"):
    rows = []
    for keys, g in df_in.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)
        rows.append((*keys, weighted_share(g[value_col], g[weight_col])))
    return pd.DataFrame(rows, columns=list(group_cols) + ["weighted_share"])

def grouped_weighted_median(df_in, group_cols, value_col, weight_col="PWSSWGT"):
    rows = []
    for keys, g in df_in.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)
        rows.append((*keys, weighted_quantile(g[value_col], 0.5, g[weight_col])))
    return pd.DataFrame(rows, columns=list(group_cols) + ["weighted_median"])
```

```python
# Plot 1: Age distribution + home internet access by metro status
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: age distribution
axes[0].hist(df_eda["age"].dropna(), bins=15, edgecolor="black", alpha=0.75)
axes[0].set_xlabel("Age")
axes[0].set_ylabel("Count")
axes[0].set_title("Age Distribution in the Cleaned Sample")

# Right: weighted internet-at-home share by metro status
internet_by_metro = grouped_weighted_share(df_eda, ["metro_group"], "internet_at_home")
internet_by_metro = internet_by_metro.sort_values("metro_group")

bars = axes[1].bar(internet_by_metro["metro_group"], internet_by_metro["weighted_share"])
axes[1].set_xlabel("Metro status")
axes[1].set_ylabel("Weighted share with internet at home")
axes[1].set_title("Home Internet Access by Metro Status")
axes[1].set_ylim(0, 1)

for bar, value in zip(bars, internet_by_metro["weighted_share"]):
    axes[1].text(bar.get_x() + bar.get_width()/2, value + 0.02, f"{value:.1%}", ha="center")

plt.tight_layout()
plt.show()

print("Missingness in main EDA variables:")
print(df_eda[["age", "internet_at_home", "metro_status", "labor_force_status_major", "hourly_earnings_clean"]].isna().mean().round(4))
```

The distribution appears relatively even across this range, with most age groups containing roughly **2000–2400 observations**, indicating that the sample is well balanced across working-age adults.

## Section 2: Labor-Market Outcomes and Demographic Patterns by Internet Access

In this section, we examine how household internet access is associated with employment outcomes and demographic characteristics. Specifically, we explore:

- whether respondents with **internet at home** have higher **employment rates**,  
- whether this relationship differs across **metro and non-metro areas**, and  
- whether demographic characteristics such as **age and gender composition** differ between households with and without internet access.

```python
# Plot 2: Employment rate + Demographic characteristics by internet access & metro status
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

employment_by_group = grouped_weighted_share(
    df_eda,
    ["metro_group", "internet_at_home"],
    "employed"
)

employment_plot = employment_by_group.pivot(
    index="metro_group",
    columns="internet_at_home",
    values="weighted_share"
).sort_index()

employment_plot.columns = ["No internet at home", "Internet at home"]
employment_plot.plot(kind="bar", ax=axes[0])
axes[0].set_xlabel("Metro status")
axes[0].set_ylabel("Weighted employment rate")
axes[0].set_title("Employment Rate by Internet Access and Metro Status")
axes[0].set_ylim(0, 1)
axes[0].legend(title="Broadband proxy")
axes[0].tick_params(axis="x", rotation=0)


demo_stats = df_eda.groupby(["metro_group", "internet_at_home"]).apply(
    lambda g: pd.Series({
        "mean_age": weighted_mean(g["age"], g["PWSSWGT"]),
        "share_male": weighted_share(g["sex"] == 1, g["PWSSWGT"])
    })
).reset_index()


demo_pivot = demo_stats.pivot(
    index="metro_group",
    columns="internet_at_home",
    values=["mean_age", "share_male"]
)


demo_pivot["mean_age"].plot(kind="bar", ax=axes[1], color=["skyblue", "orange"], position=0, width=0.4)
axes[1].set_ylabel("Mean age", color="black")
axes[1].tick_params(axis="y", labelcolor="black")


ax2 = axes[1].twinx()
demo_pivot["share_male"].plot(kind="bar", ax=ax2, color=["lightgreen", "red"], position=1, width=0.4)
ax2.set_ylabel("Share male", color="black")
ax2.tick_params(axis="y", labelcolor="black")

axes[1].set_xlabel("Metro status")
axes[1].set_title("Demographic Characteristics by Internet Access & Metro Status")
axes[1].legend(["Mean age (No internet)", "Mean age (Internet)", "Share male (No internet)", "Share male (Internet)"], 
               bbox_to_anchor=(1.15, 1), loc="upper left")
axes[1].tick_params(axis="x", rotation=0)

plt.tight_layout()
plt.show()
```

The left panel compares weighted employment rates across metro status and internet access groups. 

In metro / urban areas, the employment rate is roughly **80% for respondents without home internet** and about **76% for those with internet access**. In non-metro / rural areas, employment rates are slightly lower overall, at approximately **74% for those without internet** and **75% for those with internet access**.

Overall, the differences between internet access groups are relatively small. While employment rates appear slightly higher in metro areas compared with non-metro areas, the presence of home internet does not show a strong or consistent association with employment in this descriptive comparison.

The right panel provides demographic context by comparing **mean age and the share of male respondents** across the same groups. The average age across groups appears very similar, roughly around the **late-30s to about 40 years old**, and the share of male respondents is also close to **about 50% across all groups**. 

Because the demographic composition is relatively similar across groups, large demographic differences are unlikely to explain the small employment differences observed in the left panel.

```python
# Plot 3: Employment by sex, internet access, and metro status
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

employment_by_sex = grouped_weighted_share(
    df_eda,
    ["sex", "internet_at_home"],
    "employed"
)
employment_sex_plot = employment_by_sex.pivot(
    index="sex",
    columns="internet_at_home",
    values="weighted_share"
)
employment_sex_plot.columns = ["No internet at home", "Internet at home"]
employment_sex_plot.index = ["Male", "Female"]
employment_sex_plot.plot(kind="bar", ax=axes[0])
axes[0].set_xlabel("Sex")
axes[0].set_ylabel("Weighted employment rate")
axes[0].set_title("Employment Rate by Sex and Home Internet Access")
axes[0].set_ylim(0, 1)
axes[0].legend(title="Broadband proxy")
axes[0].tick_params(axis="x", rotation=0)

summary_rows = []
for (metro_name, internet_home), g in df_eda.groupby(["metro_group", "internet_at_home"], dropna=False):
    summary_rows.append({
        "Metro group": metro_name,
        "Internet at home": "Yes" if internet_home == 1 else "No",
        "Employment rate": round(weighted_share(g["employed"], g["PWSSWGT"]), 3),
        "Mean age": round(weighted_mean(g["age"], g["PWSSWGT"]), 1)
    })

summary_table = pd.DataFrame(summary_rows).sort_values(["Metro group", "Internet at home"]).reset_index(drop=True)

axes[1].axis("off")
tbl = axes[1].table(
    cellText=summary_table.values,
    colLabels=summary_table.columns,
    loc="center",
    cellLoc="center"
)
tbl.auto_set_font_size(False)
tbl.set_fontsize(9)
tbl.scale(1.05, 1.6)
axes[1].set_title("Summary: Employment Outcomes", pad=12)

plt.tight_layout()
plt.show()
```

The left panel compares employment rates across sex and home internet access groups. Overall, employment rates are higher for males than for females in this sample. Male respondents have employment rates of roughly **0.85 without home internet** and about **0.81 with internet access**, while female respondents have lower employment rates of approximately **0.72 without internet** and **0.69 with internet**.

Within each sex group, the difference between respondents with and without home internet access appears relatively small. In fact, employment rates are slightly higher among individuals without internet access in both groups, though the differences are modest.

These results suggest that **sex differences in employment are much larger than differences associated with internet access** in this descriptive comparison. This indicates that demographic factors such as gender may play a stronger role in employment outcomes than home internet access in this sample.

The table on the right summarizes employment rates and mean age across metro and internet-access groups. Employment rates are slightly higher in metro areas (around **0.80**) than in non-metro areas (around **0.75**), while the mean age across groups is very similar, ranging from approximately **39 to 40 years old**. The similarity in age across groups suggests that age composition is unlikely to drive the employment differences observed above.

## Ethics

This project involves two main ethical considerations related to data bias and the use of sensitive information. First, selection bias is an important concern. Household broadband access is not randomly assigned. People who have broadband often differ from those who do not in ways that also affect employment and earnings, such as education level, income, job type, and where they live. These differences may be especially strong between rural and urban areas. Although we control for many demographic and human-capital variables, some unobserved differences may remain. Because of this, we interpret our results as associations rather than causal effects. Second, the use of sensitive socioeconomic variables raises privacy and ethical considerations. Variables such as employment status, earnings, immigration status, and English proficiency can be sensitive and potentially stigmatizing. While the Pew datasets are publicly available and anonymized, we avoid reporting small subgroups or detailed geographic information that could risk harm or misinterpretation. All results are presented in aggregated form. And the analysis focuses on overall patterns rather than individual outcomes.
