## Abstract

This study examines whether household Internet access is significantly associated with the probability of employment among adults aged 25 to 54 in the United States, and whether this relationship differs between metro and non-metro areas. In the context of the rapidly evolving digital economy, the Internet has become a critical component of labor market participation; however, access to and usage remain unevenly distributed. This study assesses the impact of the digital divide on employment outcomes using empirical data analysis.

Methodologically, the study utilizes the Population Survey (CPS) Internet Use Supplement released by the U.S. Census Bureau to estimate a weighted binary generalized linear model (GLM) with a logit link function. Employment status is specified as the dependent variable, with household Internet access as the primary independent variable and an interaction term for non-metropolitan residence. Key demographic covariates—including age, gender, education, race, and marital status—are controlled for to address potential confounding. Model estimates, statistical significance, and predicted probabilities are used to compare employment likelihood across groups.

The results indicate a significant positive association between household Internet access and employment probability. Moreover, this relationship varies between metro and non-metro areas, suggesting that geographic context moderates the role of Internet access in labor market outcomes. Overall, the findings highlight the importance of digital access in shaping employment opportunities and provide evidence relevant to ongoing discussions of inequality in the digital economy.

## Research Question

Among U.S. adults aged 25–54, is having internet access at home associated with a higher probability of being employed, and does this relationship differ between metro and non-metro residents?
We plan to answer this question by creating a **single weighted binomial generalized linear model (GLM) with a logit link**

**Depdendent Variable (DV):**
the dependent variable is **employment status (`labor_force_status_major`)**
- Employment (binary: employed vs not employed)

**Indepdendent Variable (IV):**

- Internet access at home (binary: yes-access (1) vs. no-access(0))
- nonmetro (metro status - binary: nonmetro (1) vs metro (0))

**Control:**
We control for key demographic and human capital characteristics, including **age, sex, education, race, and marital status**, to account for confounding differences across individuals.
- Age, sex, race/ethnicity
- Education level
- Marital status, number of children in household

## Background and Prior Work

High-speed internet access has become a key input to modern labor markets: it enables online job search and matching, remote work, access to training, and participation in digitally mediated services (e.g., gig work, online freelancing). Yet access and adoption are uneven across the United States, creating a persistent “digital divide” that often maps onto geography (rural vs. urban), income, and education. For this project, we ask whether household broadband subscription is associated with employment and earnings for U.S. adults ages 25–54, and whether the relationship differs in rural versus urban contexts after controlling for demographic and human-capital factors.

A large body of prior research suggests that broadband can affect labor outcomes, but the direction and magnitude may depend on worker characteristics and local context. Using U.S. data from 1999–2007, Atasoy (2013) finds that broadband expansion is associated with improved labor-market outcomes, consistent with the idea that connectivity reduces frictions in job search and supports labor-force participation.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

More causal work focusing on specific subpopulations also finds meaningful effects: Dettling (2017) uses an instrumental-variables strategy based on supply-side constraints to broadband access and reports that exogenous increases in high-speed internet use raise labor force participation among married women, highlighting a plausible mechanism where internet access expands feasible work arrangements and job search at home.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)

Beyond participation, broadband can also influence who benefits in the labor market. Akerman, Gaarder, and Mogstad (2015) provide evidence of “skill complementarity,” where broadband adoption improves outcomes for more-skilled workers and can disadvantage less-skilled workers—suggesting that broadband may amplify existing inequalities unless paired with complementary skills and opportunities.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)

Recent research also emphasizes that the broadband–employment relationship may look different in rural areas, where infrastructure, adoption, and job opportunities differ from urban labor markets. For example, Isley (2022) studies rural U.S. counties during the early COVID-19 period and finds that both broadband availability and adoption are related to employment rates, using a two-stage least squares approach to address endogeneity.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)

This motivates our rural–urban heterogeneity component: if broadband supports telework, job search, and access to wider labor markets, the payoff may be larger where geographic isolation is a bigger constraint; alternatively, if high-quality jobs and complementary skills are concentrated in urban areas, the earnings gains from broadband adoption may be larger in cities. Together, these studies justify examining (1) the overall association between household broadband subscription and employment/earnings and (2) whether that association differs across rural versus urban settings after accounting for education, age, race/ethnicity, household composition, and place-based factors.

Finally, prior work highlights an important methodological challenge: broadband adoption is not randomly assigned. Households with broadband may differ systematically in income, education, occupation, and local labor conditions, all of which also affect employment and earnings. The literature therefore often emphasizes careful adjustment strategies (e.g., rich controls, fixed effects, quasi-experimental instruments, or matching) to reduce confounding and clarify interpretation.<a name="cite_ref-2b"></a>[<sup>2</sup>](#cite_note-2) In this project, using public microdata, we will treat results primarily as associational unless we can justify a credible identification strategy.

## Hypothesis

We hypothesize that having internet access at home is **positively associated** with the probability of being employed among prime-age adults. Specifically, individuals with internet access at home are expected to have **higher odds of employment** than those without access, holding demographic characteristics constant.

We further hypothesize that this association differs by geography, such that the relationship between internet access and employment is **not identical in metro and non-metro areas**, which is tested using an interaction between internet access and metro status.

In this project, “positively associated” is defined as:
- an **odds ratio greater than 1** on the internet access variable in the binomial GLM, and  
- a **statistically significant coefficient** (p-value < 0.05) after controlling for covariates.

To aid interpretation, we also evaluate **predicted probabilities of employment** across the four key groups:
- metro with and without internet access, and  
- non-metro with and without internet access.

The null hypothesis is that internet access has **no association** with employment probability and that there is **no difference** in this relationship between metro and non-metro residents.

## Shortcomings

This dataset has several limitations:

1. **Self-report bias**  
   Internet usage behavior and income information are self-reported by respondents, which may introduce measurement errors.

2. **Limited population coverage for some variables**  
   Some variables apply only to specific groups (for example, hourly earnings only apply to employed individuals), which may lead to missing or special coded values.

3. **Cross-sectional data**  
   The dataset reflects information from a single point in time, which limits the ability to infer causal relationships.

4. **Internet usage variables are categorical**  
   Internet usage variables mainly indicate whether the internet is used in certain locations rather than precise usage time or frequency.

```python
# Prepare analysis copy
df_eda = df.copy()

# Keep the variables needed for the main EDA section
key_cols = [
    "age", "internet_at_home", "metro_status", "labor_force_status_major",
    "hourly_earnings", "PWSSWGT", "sex", "education", "race", "marital_status"
]
print("Key analysis columns present:")
print([c for c in key_cols if c in df_eda.columns])

# Metro labels used throughout the EDA
metro_map = {1: "Metro / urban", 2: "Non-metro / rural"}
df_eda = df_eda[df_eda["metro_status"].isin([1, 2])].copy()
df_eda["metro_group"] = df_eda["metro_status"].map(metro_map)

# Broadband proxy
df_eda = df_eda[df_eda["internet_at_home"].isin([0, 1])].copy()

# Employment indicator
df_eda = df_eda[df_eda["labor_force_status_major"] != -1].copy()
df_eda["employed"] = df_eda["labor_force_status_major"].isin([1, 2]).astype(int)

# Clean hourly earnings
df_eda["hourly_earnings_clean"] = pd.to_numeric(df_eda["hourly_earnings"], errors="coerce")
df_eda.loc[df_eda["hourly_earnings_clean"] <= 0, "hourly_earnings_clean"] = np.nan

df_eda["log_hourly_earnings"] = np.log(
    df_eda["hourly_earnings_clean"].clip(lower=1)
)

print("\nAnalysis sample shape:", df_eda.shape)
print("Age range:", int(df_eda["age"].min()), "to", int(df_eda["age"].max()))
print("Metro groups:", df_eda["metro_group"].value_counts().to_dict())
```

```python
# Weighted helper functions
def weighted_mean(x, w):
    x = np.asarray(x, dtype=float)
    w = np.asarray(w, dtype=float)
    m = np.isfinite(x) & np.isfinite(w)
    if m.sum() == 0 or w[m].sum() == 0:
        return np.nan
    return (x[m] * w[m]).sum() / w[m].sum()

def weighted_share(x, w):
    return weighted_mean(x, w)

def weighted_quantile(values, quantile, sample_weight):
    values = np.asarray(values, dtype=float)
    sample_weight = np.asarray(sample_weight, dtype=float)
    m = np.isfinite(values) & np.isfinite(sample_weight)
    values = values[m]
    sample_weight = sample_weight[m]
    if len(values) == 0 or sample_weight.sum() == 0:
        return np.nan
    sorter = np.argsort(values)
    values = values[sorter]
    sample_weight = sample_weight[sorter]
    cum_w = np.cumsum(sample_weight)
    cutoff = quantile * sample_weight.sum()
    return values[np.searchsorted(cum_w, cutoff, side="left")]

def grouped_weighted_share(df_in, group_cols, value_col, weight_col="PWSSWGT"):
    rows = []
    for keys, g in df_in.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)
        rows.append((*keys, weighted_share(g[value_col], g[weight_col])))
    return pd.DataFrame(rows, columns=list(group_cols) + ["weighted_share"])

def grouped_weighted_median(df_in, group_cols, value_col, weight_col="PWSSWGT"):
    rows = []
    for keys, g in df_in.groupby(group_cols, dropna=False):
        if not isinstance(keys, tuple):
            keys = (keys,)
        rows.append((*keys, weighted_quantile(g[value_col], 0.5, g[weight_col])))
    return pd.DataFrame(rows, columns=list(group_cols) + ["weighted_median"])
```

### Section 2: Labor-Market Outcomes and Demographic Patterns by Internet Access

- whether this relationship differs across **metro and non-metro areas**, and

```python
# Plot 2: Employment rate + Demographic characteristics by internet access & metro status
fig, ax = plt.subplots(1, 1, figsize=(8, 5))

# Plot weighted employment rates across metro status and internet access
employment_by_group = grouped_weighted_share(
    df_eda,
    ["metro_group", "internet_at_home"],
    "employed"
)
employment_plot = employment_by_group.pivot(
    index="metro_group",
    columns="internet_at_home",
    values="weighted_share"
).sort_index()
employment_plot.columns = ["No internet at home", "Internet at home"]
employment_plot.plot(kind="bar", ax=ax)
ax.set_xlabel("Metro status")
ax.set_ylabel("Weighted employment rate")
ax.set_title("Employment Rate by Internet Access and Metro Status")
ax.set_ylim(0, 1)
ax.legend(title="Broadband proxy")
ax.tick_params(axis="x", rotation=0)

plt.tight_layout()
plt.show()

# Plot mean age to examine demographic differences across groups
demo_stats = df_eda.groupby(["metro_group", "internet_at_home"]).apply(
    lambda g: pd.Series({
        "mean_age": weighted_mean(g["age"], g["PWSSWGT"]),
        "share_male": weighted_share(g["sex"] == 1, g["PWSSWGT"])
    }),
    include_groups=False
).reset_index()

demo_pivot = demo_stats.pivot(
    index="metro_group",
    columns="internet_at_home",
    values=["mean_age", "share_male"])


fig, (ax_age, ax_male) = plt.subplots(1, 2, figsize=(12, 5))

demo_pivot["mean_age"].plot(
    kind="bar",
    ax=ax_age,
    color=["skyblue", "orange"],
    width=0.6
)
ax_age.set_ylabel("Mean age")
ax_age.set_title("Mean Age by Metro Status & Internet Access")
ax_age.legend(title="Internet at home", labels=["No", "Yes"])
ax_age.set_ylim(38, 41)
ax_age.tick_params(axis="x", rotation=0)

# Plot gender composition (share of male respondents) across groups
demo_pivot["share_male"].plot(
    kind="bar",
    ax=ax_male,
    color=["skyblue", "orange"],
    width=0.6
)
ax_male.set_ylabel("Share male")
ax_male.set_title("Share of Male Respondents by Metro Status & Internet Access")
ax_male.legend(title="Internet at home", labels=["No", "Yes"])
ax_male.set_ylim(0.45, 0.55)
ax_male.tick_params(axis="x", rotation=0)

plt.tight_layout()
plt.show()
```

We proceed to a regression analysis to estimate the association between internet access and employment outcomes while controlling for key covariates and testing for interaction effects.

## Analysis 1: Binormial GML with a logit link

Our research question focus on exploring whether **internet access at home** is associated with **employment** for adults ages 25–54, and whether that association differs between **metro and non-metro** residents. 

A **binomial generalized linear model (GLM) with a logit link** is the best fit for this hypothesis test for three reasons.

First, the dependent variable, `employed`, is **binary**. Each respondent is either coded as employed or not employed. A binomial GLM is designed for exactly this kind of outcome, whereas ordinary least squares is not ideal because it can predict values outside the 0–1 range and assumes constant error variance.

Second, the hypothesis is about whether the probability of employment changes with `internet_at_home`, and whether that relationship differs by `nonmetro` status. A logit-link binomial GLM models the relationship between predictors and the **log-odds of employment**, which is appropriate when the underlying quantity of interest is a probability bounded between 0 and 1.

Third, we uses CPS survey weights (`PWSSWGT`). The GLM implementation lets us include those weights directly, which makes the analysis better aligned with the survey structure than an unweighted model would be.

To match the testing intention of the hypothesis, the model includes:

- `internet_at_home`
- `nonmetro`
- `internet_at_home × nonmetro`
- age and age squared
- sex, education, race, and marital status

The interaction term is the key test of whether the internet–employment association differs between metro and non-metro residents.


```python
analysis_df = df_eda.copy()
analysis_df["nonmetro"] = (analysis_df["metro_status"] != 1).astype(int)
analysis_df["employed"] = (analysis_df["labor_force_status_major"] == 1).astype(int)
# Main analysis sample:
# keep observations needed for the employment model and the exact variables from the research question
model_df = analysis_df.dropna(subset=[
    "internet_at_home", "nonmetro", "age", "sex", "education", "race", "marital_status", "PWSSWGT", "employed"
]).copy()

# Make sure the main explanatory variable is binary and interpretable
model_df = model_df[model_df["internet_at_home"].isin([0, 1])].copy()

main_formula = (
    "employed ~ internet_at_home * nonmetro"
    " + age + I(age**2)"
    " + C(sex) + C(education) + C(race) + C(marital_status)"
)

main_glm = smf.glm(
    formula=main_formula,
    data=model_df,
    family=sm.families.Binomial(),
    freq_weights=model_df["PWSSWGT"]
).fit(cov_type="HC3")

main_or = pd.DataFrame({
    "odds_ratio": np.exp(main_glm.params),
    "ci_low": np.exp(main_glm.conf_int()[0]),
    "ci_high": np.exp(main_glm.conf_int()[1]),
    "p_value": main_glm.pvalues
})

focus_terms = ["internet_at_home", "nonmetro", "internet_at_home:nonmetro"]
main_or.loc[focus_terms].round(3)
```

### Multicollinearity Check (VIF)

To ensure that the model coefficients are stable and interpretable, we check for multicollinearity using the Variance Inflation Factor (VIF). High VIF values (commonly above 5–10) indicate problematic collinearity.

This is a diagnostic step and does not change the model—it only helps assess whether the predictors are excessively correlated.


```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Select key numeric predictors for VIF check
vif_df = model_df[["internet_at_home", "nonmetro", "age"]].copy()
vif_df["age_sq"] = vif_df["age"]**2

# Add constant
vif_df = sm.add_constant(vif_df)

vif_results = pd.DataFrame({
    "feature": vif_df.columns,
    "VIF": [variance_inflation_factor(vif_df.values, i) for i in range(vif_df.shape[1])]
})

vif_results
```

The VIF results show that age and age_sq have very high VIF values, which is expected because age_sq is a deterministic transformation of age. This type of multicollinearity is not problematic for interpretation, as both terms are included intentionally to capture a nonlinear relationship.

Importantly, the other predictors (internet_at_home and nonmetro) have VIF values close to 1, indicating no concerning multicollinearity among the main variables of interest. Therefore, multicollinearity is not a major concern for the key relationships in this model.


```python
# Predicted probabilities for the four core groups in the research question

controls = {
    "age": model_df["age"].median(),
    "sex": model_df["sex"].mode()[0],
    "education": model_df["education"].mode()[0],
    "race": model_df["race"].mode()[0],
    "marital_status": model_df["marital_status"].mode()[0],
}

pred_main = pd.DataFrame([
    {"internet_at_home": 0, "nonmetro": 0, **controls},
    {"internet_at_home": 1, "nonmetro": 0, **controls},
    {"internet_at_home": 0, "nonmetro": 1, **controls},
    {"internet_at_home": 1, "nonmetro": 1, **controls},
])

pred_main["metro_group"] = np.where(pred_main["nonmetro"] == 1, "Non-metro / rural", "Metro / urban")
pred_main["internet_label"] = np.where(pred_main["internet_at_home"] == 1, "Internet at home", "No internet at home")
pred_main["predicted_employment_probability"] = main_glm.predict(pred_main)

plt.figure(figsize=(8, 5))
sns.barplot(data=pred_main, x="metro_group", y="predicted_employment_probability", hue="internet_label")
plt.xlabel("Metro status")
plt.ylabel("Predicted probability of employment")
plt.title("Adjusted Employment Predictions from the Main Model")
plt.ylim(0, 1)
plt.legend(title="")
plt.tight_layout()
plt.show()

pred_main[["metro_group", "internet_label", "predicted_employment_probability"]].round(3)
```

**Interpretation of the main model**

The main binomial GLM shows that the association between home internet access and employment depends on geography.

- In **metro areas** (the reference group), the odds ratio for `internet_at_home` is **0.865** (*p* < 0.001). This means that, after controlling for age, sex, education, race, and marital status, respondents with home internet access have about **13.5% lower odds of employment** than otherwise similar metro respondents without home internet access.
- The odds ratio for `nonmetro` is **0.831** (*p* < 0.001). Among respondents without home internet access, non-metro residents have about **16.9% lower odds of employment** than comparable metro residents.
- The interaction term, `internet_at_home × nonmetro`, has an odds ratio of **1.231** (*p* < 0.001). This indicates that the internet-at-home association is **more positive in non-metro areas** than in metro areas.

To make that interaction easier to interpret, we look at the predicted probabilities. Holding the control variables constant:
- in **metro areas**, predicted employment falls from **82.6%** to **80.4%** when moving from no home internet access to home internet access, a decrease of about **2.2 percentage points**;
- in **non-metro areas**, predicted employment rises from **79.8%** to **80.8%**, an increase of about **1.0 percentage point**.

So the main model does **not** support the original hypothesis that home internet access is positively associated with employment everywhere. Instead, it suggests a **negative association in metro areas** and a **small positive association in non-metro areas**.


## Analysis 2: Robustness Check Using an Alternative Internet Measure

To ensure that our research results are strong and important we perform a check on whether our main result depends too heavily on one operationalization of digital access. The cleaned file contains a second internet-related variable, `internet_user`, which captures whether the respondent is coded as an internet user more broadly. Although it is not identical to home internet access, it is still a reasonable sensitivity check because it measures the same broader concept of digital connectedness.

This robustness model uses the **same binomial GLM framework** as the main analysis. That consistency matters: if the model form changed at the same time as the internet variable, it would be harder to tell whether any differences came from the variable definition or from the method itself. By keeping the outcome, link function, interaction structure, and controls the same, the only substantive change is the internet measure.


```python
robust_df = analysis_df.dropna(subset=[
    "internet_user", "nonmetro", "age", "sex", "education", "race", "marital_status", "PWSSWGT", "employed"
]).copy()

robust_df = robust_df[robust_df["internet_user"].isin([0, 1])].copy()

robust_formula = (
    "employed ~ internet_user * nonmetro"
    " + age + I(age**2)"
    " + C(sex) + C(education) + C(race) + C(marital_status)"
)

robust_glm = smf.glm(
    formula=robust_formula,
    data=robust_df,
    family=sm.families.Binomial(),
    freq_weights=robust_df["PWSSWGT"]
).fit(cov_type="HC3")

robust_or = pd.DataFrame({
    "odds_ratio": np.exp(robust_glm.params),
    "ci_low": np.exp(robust_glm.conf_int()[0]),
    "ci_high": np.exp(robust_glm.conf_int()[1]),
    "p_value": robust_glm.pvalues
})

comparison = pd.DataFrame({
    "Main model OR": main_or.loc[["internet_at_home", "nonmetro", "internet_at_home:nonmetro"], "odds_ratio"].values,
    "Main model p": main_or.loc[["internet_at_home", "nonmetro", "internet_at_home:nonmetro"], "p_value"].values,
    "Robustness OR": robust_or.loc[["internet_user", "nonmetro", "internet_user:nonmetro"], "odds_ratio"].values,
    "Robustness p": robust_or.loc[["internet_user", "nonmetro", "internet_user:nonmetro"], "p_value"].values,
}, index=["internet measure", "nonmetro", "interaction"])

comparison.round(3)
```

The robustness model tells essentially the same story as the main model. Replacing `internet_at_home` with `internet_user` produces very similar estimates:

- main internet coefficient: **0.865** in the main model versus **0.853** in the robustness model,
- non-metro coefficient: **0.831** versus **0.829**,
- interaction term: **1.231** versus **1.218**.

All three focal terms remain statistically significant (*p* < 0.001), and the signs do not change. This means the central pattern is stable across both digital-access measures: the internet coefficient is negative in metro areas, while the interaction term is positive, making the association less negative—and slightly positive in probability terms—in non-metro areas.

Because the robustness model reaches the same substantive conclusion, the unexpected direction of the metro-area result does not appear to be caused only by the specific choice of `internet_at_home` as the digital-access measure.


```python
# Robustness predicted probabilities

pred_robust = pd.DataFrame([
    {"internet_user": 0, "nonmetro": 0, **controls},
    {"internet_user": 1, "nonmetro": 0, **controls},
    {"internet_user": 0, "nonmetro": 1, **controls},
    {"internet_user": 1, "nonmetro": 1, **controls},
])

pred_robust["metro_group"] = np.where(pred_robust["nonmetro"] == 1, "Non-metro / rural", "Metro / urban")
pred_robust["internet_label"] = np.where(pred_robust["internet_user"] == 1, "Internet user", "Not coded as internet user")
pred_robust["predicted_employment_probability"] = robust_glm.predict(pred_robust)

plt.figure(figsize=(8, 5))
sns.barplot(data=pred_robust, x="metro_group", y="predicted_employment_probability", hue="internet_label")
plt.xlabel("Metro status")
plt.ylabel("Predicted probability of employment")
plt.title("Adjusted Employment Predictions Using the Alternative Internet Measure")
plt.ylim(0, 1)
plt.legend(title="")
plt.tight_layout()
plt.show()

pred_robust[["metro_group", "internet_label", "predicted_employment_probability"]].round(3)
```

This robustness check supports the same substantive interpretation as the main analysis rather than pointing to a different conclusion. That is important because it shows that the project's central finding is **stable across two related internet-use measures**.

In other words:
- the result is **not** that internet access is uniformly associated with higher employment;
- instead, the result is that the association differs by geography, with a **negative metro-area association** and a **more favorable non-metro association**.

That makes this robustness analysis a good and necessary follow-up to the main model because it tests the same hypothesis structure rather than introducing a separate question.


## Analysis 3: Model Interpretation and Hypothesis Evaluation

After estimating the main binomial GLM and the robustness model, the final step is to connect the statistical output back to the hypothesis in a direct and readable way. This section does not introduce a new method. Instead, it organizes the results around the exact questions the hypothesis asked:

1. Is internet access at home associated with employment?
2. Is that association different in non-metro areas?
3. Does the conclusion look similar when using a broader internet-use measure?

This is an appropriate final analysis step because good hypothesis testing is not just about running a model. It is also about explaining how the model answers the question and what the fitted results mean substantively.


```python
hypothesis_summary = pd.DataFrame({
    "Question": [
        "Internet-at-home association in metro areas",
        "Non-metro baseline difference",
        "Difference in slope for non-metro areas",
        "Direction stable in robustness model?"
    ],
    "Evidence": [
        f"OR = {main_or.loc['internet_at_home', 'odds_ratio']:.3f}, p < 0.001",
        f"OR = {main_or.loc['nonmetro', 'odds_ratio']:.3f}, p < 0.001",
        f"OR = {main_or.loc['internet_at_home:nonmetro', 'odds_ratio']:.3f}, p < 0.001",
        
        "Yes" if (
            (main_or.loc['internet_at_home', 'odds_ratio'] > 1 and robust_or.loc['internet_user', 'odds_ratio'] > 1) or
            (main_or.loc['internet_at_home', 'odds_ratio'] < 1 and robust_or.loc['internet_user', 'odds_ratio'] < 1)
        ) else "No"
    ]
})

hypothesis_summary
```

Taken together, the analysis section is tightly matched to the research question and hypothesis. Every formal model below the EDA serves one of two purposes:

- **directly testing whether internet access at home is associated with employment**, and whether that association differs by metro status, or
- **checking whether that conclusion is robust to a related internet-use measure**.

The binomial GLM is especially appropriate here because the hypothesis is about differences in a **binary labor-market outcome**. Its fitted coefficients are most naturally interpreted through **odds ratios** and then translated into **predicted probabilities** for the four groups named in the hypothesis. That makes the analysis both statistically appropriate and substantively readable.


The hypothesis-evaluation summary points to a mixed result. The association between home internet access and employment is **not positive overall in the way originally hypothesized**. In metro areas, the odds ratio for home internet access is **0.865** (*p* < 0.001), indicating lower adjusted employment odds for respondents with home internet access. However, the interaction term is **1.231** (*p* < 0.001), which means the relationship becomes more positive in non-metro areas.

Combining those two pieces, the non-metro internet-at-home association is more favorable than the metro association and is modestly positive in the predicted probabilities. That is why the results are best described as **geographically heterogeneous** rather than simply positive or negative.

The robustness model reinforces that interpretation. Using `internet_user` instead of `internet_at_home` leads to nearly identical conclusions, so the main result appears stable across both measures of digital access.

## Ethics

This project raises important ethical issues about **dataset bias, measurement limitations, and responsible interpretation of statistical results**.

First, internet access is not randomly assigned to the population. People with and without internet access differ in education, household resources, caregiving responsibilities, disability status, local labor demand, and many other structural factors that may also affect employment outcomes. Because of these differences, the dataset may reflect selection bias and omitted-variable bias. Even after adding demographic controls, some relevant factors may still be unobserved. For that reason, the models in this report should be interpreted as associational rather than causal. Recognizing these sources of dataset bias helps prevent misleading conclusions about the relationship between internet access and employment.

Second, there are also measurement limitations in the data. The project originally aimed to analyze both employment and earnings, but the processed hourly_earnings variable in the supplied cleaned file is unusable because it has no meaningful variation after cleaning. In addition, internet_at_home and internet_user are proxies for digital access rather than direct validated measures of broadband quality, speed, or reliability. These variables capture whether someone reports internet access, but they do not fully represent the complexity of digital connectivity. Ethically, it is important to acknowledge these measurement limitations and avoid overstating the strength or precision of the results.

Third, the project includes demographic controls such as race, sex, education, and marital status to reduce confounding differences across individuals. These variables are used to improve statistical adjustment rather than to rank or stereotype groups. A responsible interpretation means recognizing that differences in the results may reflect broader social and economic conditions rather than inherent characteristics of demographic groups. For this reason, our visualizations, summary statistics, and regression results are presented carefully. We aim to describe what the data show without overstating the results or implying causal relationships between internet access and employment.

Finally, we clearly acknowledge the limitations of both the dataset and the model used in this analysis. As discussed above, the data may contain some bias, and several variables are only proxies for the concepts we are trying to measure. Because of this, we interpret the results carefully and avoid making causal claims. Instead, the findings should be understood as descriptive associations that reflect the limitations of the data and the modeling approach.
